### Data Views

```python
df = pd.read_csv('titanic/train.csv')
df.info()
df
```

```python
df = df.dropna(subset=['Age', 'Embarked'])
df = df.fillna({'Cabin': 'Unknown'})

df['Age'] = df['Age'].apply(lambda x: x//1 + 1.0 if x % 1 > 0.0 else x)
```

```python
# make dummy variables for categorical columns
categorical_columns = ['Embarked', 'Cabin']
```

```python

df['Age'] = df['Age'].astype(int)
categorical_columns = ['Embarked', 'Cabin']
df['Embarked'] = df['Embarked'].astype('category')
```

```python
df.to_csv('titanic/train_cleaned.csv', index=False)
```
